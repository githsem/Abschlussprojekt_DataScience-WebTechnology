# Textblob Klassifikation

```python
import textblob
import pandas as pd
import numpy as np
from textblob_de import TextBlobDE
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
df = pd.read_csv("kbClean.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
reviews = [str(a) for a in df['Erfahrungsbericht']]
resultPolarity = []
```

```python
for review in reviews:
    bewertungen = TextBlobDE(review)
    resultPolarity.append(bewertungen.polarity)
```

```python
len(resultPolarity)
```

```python
resultPolarity
```

```python
df['Polarity Level'] = resultPolarity
```

```python
df.round(2).head(50)
```

```python
# df.groupby('NameKlinik').plot.bar();
```

```python
final_df = df.groupby(['NameKlinik'])['Polarity Level'].mean()
```
