```python
import textblob
from textblob_de import TextBlobDE
import pandas as pd

```

```python
df = pd.read_csv("../GoogleMaps_Scraping/GoogleMaps.csv")
```

```python
df = df.replace('\n',' ', regex=True)
```

```python
df.head()
```

```python
bewertungen = [str(a) for a in df['Bewertung']]
ergebnisse = []
```

```python
for b in bewertungen:
    a = TextBlobDE(b)
    ergebnisse.append(a.polarity)
```

```python
ergebnisse[:20]
```

```python
len(ergebnisse)
```

```python
df['Polarity']=ergebnisse
```
